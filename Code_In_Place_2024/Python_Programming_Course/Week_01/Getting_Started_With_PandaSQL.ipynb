```python
AVG_data = pysqldf("""SELECT nationality, COUNT(sofifa_id) AS Total_nationality
FROM data
GROUP BY nationality
HAVING COUNT(sofifa_id) > 2""")
AVG_data
```

```python
Top_5=pysqldf("""SELECT * FROM AVG_data
ORDER BY Total_nationality DESC
LIMIT 5""")
Top_5
```

```python
import matplotlib.pyplot as plt
import numpy as np
values=np.array(Top_5["Total_nationality"])
mylabels =Top_5["nationality"]
```

```python
myexplode = [0.2,0,0,0,0]
plt.title("Percentage of Top 5 Country Players ")
plt.pie(values, labels = mylabels,explode=myexplode, autopct='%1.0f%%', shadow = True)
plt.legend(title = "Players By Country", loc='lower left')
plt.show() 
```

```python
top_5_wages=pysqldf("""SELECT
 nationality,
  ROUND(AVG(wage_eur),0) as Wages_Avg
FROM data
GROUP BY club_name
HAVING AVG(wage_eur)>= 60000 ORDER BY Wages_Avg DESC""")
top_5_wages
```

```python
plt.bar(top_5_wages.nationality,top_5_wages.Wages_Avg, color="green")

plt.title("Bar Graph")
plt.xlabel("Region")
plt.ylabel("Crime Rate")
plt.show()
```
